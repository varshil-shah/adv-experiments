```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
from datetime import datetime

# Load the dataset
file_path = '/content/healthcare_dataset.csv'
data = pd.read_csv(file_path)

# Convert 'Date of Admission' and 'Discharge Date' to datetime format for time-based analysis
data['Date of Admission'] = pd.to_datetime(data['Date of Admission'])
data['Discharge Date'] = pd.to_datetime(data['Discharge Date'])
```

```python
# Word Chart: Frequency distribution of 'Medical Condition'
plt.figure(figsize=(10, 6))
sns.countplot(data=data, y='Medical Condition', order=data['Medical Condition'].value_counts().index, palette='viridis')
plt.title('Frequency Distribution of Medical Conditions')
plt.show()
```

```python
# Box and Whisker Plot: Distribution of 'Billing Amount' by 'Admission Type'
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Admission Type', y='Billing Amount', palette='Set2')
plt.title('Billing Amount Distribution by Admission Type')
plt.show()
```

```python
# Violin Plot: Distribution of 'Billing Amount' by 'Gender'
plt.figure(figsize=(10, 6))
sns.violinplot(data=data, x='Gender', y='Billing Amount', palette='muted')
plt.title('Billing Amount Distribution by Gender')
plt.show()
```

```python
# Regression Plot: Relation between 'Age' and 'Billing Amount' (linear and nonlinear)
plt.figure(figsize=(10, 6))
sns.regplot(data=data, x='Age', y='Billing Amount', scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.title('Linear Regression: Age vs. Billing Amount')
plt.show()
```

```python
# 3D Chart: Relationship between 'Age', 'Billing Amount', and 'Room Number'
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['Age'], data['Billing Amount'], data['Room Number'], c='r', marker='o')
ax.set_xlabel('Age')
ax.set_ylabel('Billing Amount')
ax.set_zlabel('Room Number')
plt.title('3D Scatter Plot: Age, Billing Amount, and Room Number')
plt.show()
```

```python
# Jitter Plot: Age vs. Billing Amount by Gender
plt.figure(figsize=(10, 6))
sns.stripplot(data=data, x='Gender', y='Billing Amount', jitter=True, palette='Set1')
plt.title('Jitter Plot: Billing Amount by Gender')
plt.show()
```

```python
# Line Plot: Trend of admissions over time
plt.figure(figsize=(10, 6))
admissions_trend = data['Date of Admission'].value_counts().sort_index()
admissions_trend.plot(kind='line', color='purple')
plt.title('Admissions Trend Over Time')
plt.xlabel('Date of Admission')
plt.ylabel('Number of Admissions')
plt.show()
```

```python

# Waterfall Plot: Cumulative distribution of Billing Amount
plt.figure(figsize=(10, 6))
data_sorted = data.sort_values('Billing Amount', ascending=False)
cumulative_billing = data_sorted['Billing Amount'].cumsum()
plt.bar(range(len(cumulative_billing)), cumulative_billing, color='orange')
plt.title('Waterfall Plot: Cumulative Billing Amount')
plt.xlabel('Records')
plt.ylabel('Cumulative Billing Amount')
plt.show()
```

```python


# Donut Chart: Proportion of different Medical Condition types
plt.figure(figsize=(8, 8))
condition_counts = data['Medical Condition'].value_counts()
plt.pie(condition_counts, labels=condition_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set3', len(condition_counts)))
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title('Proportion of Medical Conditions')
plt.show()
```

```python
# Funnel Plot: Billing amount funnel for different Admission Type
plt.figure(figsize=(10, 6))
funnel_data = data.groupby('Admission Type')['Billing Amount'].sum().sort_values(ascending=False)
plt.barh(funnel_data.index, funnel_data, color='dodgerblue')
plt.title('Funnel Plot: Billing Amount by Admission Type')
plt.xlabel('Total Billing Amount')
plt.ylabel('Admission Type')
plt.show()
```
